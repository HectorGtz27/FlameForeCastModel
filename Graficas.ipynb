```python
import matplotlib.pyplot as plt
import numpy as np

# Tamao del encuadre
size = 64

# Crear una matriz 2D de ceros para representar el encuadre
frame_before = np.zeros((size, size))
frame_after = np.zeros((size, size))

# Definir la posicin inicial de la trayectoria antes y despus de la prediccin
x_before, y_before = size // 2, size // 2
x_after, y_after = size // 2, size // 2

# Simular una trayectoria suave con una distribucin gaussiana antes de la prediccin
for _ in range(1000):  # Nmero de pasos en la trayectoria
    # Generar desplazamientos aleatorios
    dx, dy = np.random.normal(0, 2), np.random.normal(0, 2)

    # Actualizar la posicin antes de la prediccin
    x_before += int(dx)
    y_before += int(dy)

    # Asegurarse de que la posicin est dentro del encuadre
    x_before = np.clip(x_before, 0, size - 1)
    y_before = np.clip(y_before, 0, size - 1)

    # Incrementar la densidad en la posicin
    frame_before[y_before, x_before] += 1

# Normalizar la matriz antes de la prediccin para ajustar los valores entre 0 y 1
frame_before /= frame_before.max()

# Simular una trayectoria suave con una distribucin gaussiana despus de la prediccin
for _ in range(1000):  # Nmero de pasos en la trayectoria
    # Generar desplazamientos aleatorios
    dx, dy = np.random.normal(0, 2), np.random.normal(0, 2)

    # Actualizar la posicin despus de la prediccin
    x_after += int(dx)
    y_after += int(dy)

    # Asegurarse de que la posicin est dentro del encuadre
    x_after = np.clip(x_after, 0, size - 1)
    y_after = np.clip(y_after, 0, size - 1)

    # Incrementar la densidad en la posicin
    frame_after[y_after, x_after] += 1

# Normalizar la matriz despus de la prediccin para ajustar los valores entre 0 y 1
frame_after /= frame_after.max()

# Crear un mapa de calor para representar la trayectoria antes de la prediccin
plt.subplot(121)
plt.imshow(frame_before, cmap='viridis', vmin=0, vmax=1)  # Cambiar la paleta de color
plt.title('Todays Fire')
cbar_before = plt.colorbar(shrink=0.7)  # Ajustar el tamao de la barra de colores
cbar_before.set_label('Intensidad de Fuego')

# Crear un mapa de calor para representar la trayectoria despus de la prediccin
plt.subplot(122)
plt.imshow(frame_after, cmap='viridis', vmin=0, vmax=1)  # Cambiar la paleta de color
plt.title('Tomorrows Fire')
cbar_after = plt.colorbar(shrink=0.7)  # Ajustar el tamao de la barra de colores
cbar_after.set_label('Intensidad de Fuego')

# Ajustar el tamao de la figura y el espaciado entre subplots
plt.gcf().set_size_inches(10, 5)
plt.subplots_adjust(wspace=0.4)  # Ajustar el espacio horizontal entre subplots

plt.show()




```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generar datos aleatorios (reemplzalos con tus datos reales)
data = {
    'Temperatura': np.random.rand(100),
    'Humedad': np.random.rand(100),
    'Viento': np.random.rand(100),
    'Intensidad del Fuego': np.random.rand(100),
}

# Crear un DataFrame de pandas con los datos
df = pd.DataFrame(data)

# Calcular la matriz de correlacin
correlation_matrix = df.corr()

# Crear una matriz de calor (heatmap) para visualizar la correlacin
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlacin de Datos Relacionados con el Fuego')
plt.show()
```
