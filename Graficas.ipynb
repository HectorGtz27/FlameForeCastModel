```python
import matplotlib.pyplot as plt
import numpy as np

# Tamaño del encuadre
size = 64

# Crear una matriz 2D de ceros para representar el encuadre
frame_before = np.zeros((size, size))
frame_after = np.zeros((size, size))

# Definir la posición inicial de la trayectoria antes y después de la predicción
x_before, y_before = size // 2, size // 2
x_after, y_after = size // 2, size // 2

# Simular una trayectoria suave con una distribución gaussiana antes de la predicción
for _ in range(1000):  # Número de pasos en la trayectoria
    # Generar desplazamientos aleatorios
    dx, dy = np.random.normal(0, 2), np.random.normal(0, 2)

    # Actualizar la posición antes de la predicción
    x_before += int(dx)
    y_before += int(dy)

    # Asegurarse de que la posición esté dentro del encuadre
    x_before = np.clip(x_before, 0, size - 1)
    y_before = np.clip(y_before, 0, size - 1)

    # Incrementar la densidad en la posición
    frame_before[y_before, x_before] += 1

# Normalizar la matriz antes de la predicción para ajustar los valores entre 0 y 1
frame_before /= frame_before.max()

# Simular una trayectoria suave con una distribución gaussiana después de la predicción
for _ in range(1000):  # Número de pasos en la trayectoria
    # Generar desplazamientos aleatorios
    dx, dy = np.random.normal(0, 2), np.random.normal(0, 2)

    # Actualizar la posición después de la predicción
    x_after += int(dx)
    y_after += int(dy)

    # Asegurarse de que la posición esté dentro del encuadre
    x_after = np.clip(x_after, 0, size - 1)
    y_after = np.clip(y_after, 0, size - 1)

    # Incrementar la densidad en la posición
    frame_after[y_after, x_after] += 1

# Normalizar la matriz después de la predicción para ajustar los valores entre 0 y 1
frame_after /= frame_after.max()

# Crear un mapa de calor para representar la trayectoria antes de la predicción
plt.subplot(121)
plt.imshow(frame_before, cmap='viridis', vmin=0, vmax=1)  # Cambiar la paleta de color
plt.title('Todays Fire')
cbar_before = plt.colorbar(shrink=0.7)  # Ajustar el tamaño de la barra de colores
cbar_before.set_label('Intensidad de Fuego')

# Crear un mapa de calor para representar la trayectoria después de la predicción
plt.subplot(122)
plt.imshow(frame_after, cmap='viridis', vmin=0, vmax=1)  # Cambiar la paleta de color
plt.title('Tomorrows Fire')
cbar_after = plt.colorbar(shrink=0.7)  # Ajustar el tamaño de la barra de colores
cbar_after.set_label('Intensidad de Fuego')

# Ajustar el tamaño de la figura y el espaciado entre subplots
plt.gcf().set_size_inches(10, 5)
plt.subplots_adjust(wspace=0.4)  # Ajustar el espacio horizontal entre subplots

plt.show()




```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generar datos aleatorios (reemplázalos con tus datos reales)
data = {
    'Temperatura': np.random.rand(100),
    'Humedad': np.random.rand(100),
    'Viento': np.random.rand(100),
    'Intensidad del Fuego': np.random.rand(100),
}

# Crear un DataFrame de pandas con los datos
df = pd.DataFrame(data)

# Calcular la matriz de correlación
correlation_matrix = df.corr()

# Crear una matriz de calor (heatmap) para visualizar la correlación
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlación de Datos Relacionados con el Fuego')
plt.show()
```
